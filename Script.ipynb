```python
# Caminho para o arquivo Excel
excel_path = 'CodeSamples - Consolidado.xlsx'

# Carrega o arquivo Excel
df = pd.read_excel(excel_path)

# Verifica as primeiras linhas do DataFrame e as colunas disponíveis
print(df.head())
print("Colunas disponíveis no DataFrame:")
print(df.columns)

# Supondo que os links estão em uma coluna chamada 'html_url'
# Altere 'html_url' para o nome exato da coluna no seu arquivo
links_coluna = 'html_url'  # Ajuste para o nome correto da coluna de links
repositorios = df[links_coluna].dropna().tolist()

# Verifica se os links foram extraídos corretamente
print(f"Total de links extraídos: {len(repositorios)}")
print(repositorios)


```

```python
import xml.etree.ElementTree as et

# Caminho para o arquivo XML
postPath = 'Posts.xml'
questions = []

with open(postPath, encoding='utf-8') as posts:
    for line in posts:
        if "<?xml" in line or "<posts>" in line:
            continue
        try:
            parsed = et.fromstring(line)
        except:
            continue
        
        if parsed.get("PostTypeId") != "1":
            continue
        
        id = int(parsed.get("Id"))
        if id % 100000 == 0:
            print(id)
        
        # Verifica se qualquer URL da lista está presente no corpo da pergunta
        body = parsed.get("Body")
        if any(repo in body for repo in repositorios):
            # Extrai mais parâmetros, se necessário
            creation_date = parsed.get("CreationDate")
            score = parsed.get("Score")
            view_count = parsed.get("ViewCount")
            title = parsed.get("Title")
            
            # Armazena os dados em um dicionário
            question_data = {
                "Id": id,
                "CreationDate": creation_date,
                "Score": score,
                "ViewCount": view_count,
                "Body": body,
                "Title": title,
                "RelatedRepo": [repo for repo in repositorios if repo in body]  # Lista de repositórios relacionados
            }
            
            questions.append(question_data)

# Exemplo: Imprime as primeiras perguntas relacionadas encontradas
for q in questions[:5]:
    print(q)
```

For each post in Posts, we try to find the github referrence in Post body.

```python
with open(postPath, encoding='utf-8') as posts:
    for line in posts:
        if "<?xml" in line or "<posts>" in line:
            continue
        try:
            parsed = et.fromstring(line)
        except:
            continue
        if parsed.get("PostTypeId") != "1":
            continue
        id = int(parsed.get("Id"))
        if id % 100000 == 0:
            print(id)
            
        if "github.com/aws-samples" in parsed.get("Body") or "github.com/Azure-Samples" in parsed.get("Body") or "github.com/googlesamples" in parsed.get("Body") or "github.com/googlearchive" in parsed.get("Body") or "github.com/spring-guides" in parsed.get("Body") or "github.com/SAP-samples" in parsed.get("Body") or "github.com/spring-cloud-samples" in parsed.get("Body") or "github.com/spring-guides" in parsed.get("Body"):
            questions.append(parsed)
```

```python
# Supondo que os elementos em `questions` sejam objetos XML
questions_data = []

for question in questions:
    questions_data.append({
        "Id": question.get("Id"),
        "CreationDate": question.get("CreationDate"),
        "OwnerUserId": question.get("OwnerUserId"),
        "PostTypeId": question.get("PostTypeId"),
        "AcceptedAnswerId": question.get("AcceptedAnswerId"),
        "Body": question.get("Body")  # Adiciona mais campos se necessário
    })

# Cria um DataFrame a partir da lista de dicionários
questions_df = pd.DataFrame(questions_data)

# Salva o DataFrame como CSV
questions_df.to_csv("Questions.csv", index=False)
```

Path of code samples by framework

Filtring only questions about the selected code samples

```python
def filterSamples(samplesPath, questions):
    lines = []
    for samplePath in samplesPath["html_url"]:
        owner = samplePath.split("/")[0]
        sample = samplePath.split("/")[1]
        for question in questions:
            if sample in question.get("Body"):
                print(f"{question.get('Id')}")
                lines.append([owner, sample, question.get("Id"), question.get("CreationDate"), question.get("OwnerUserId"), question.get("PostTypeId"), question.get("AcceptedAnswerId")])
    return lines
```

```python
def createDataFrame(lines):
    columns = ["framework", "path", "id", "creationDate", "ownerUserId", "postTypeId", "acceptedAnswerId"]
    df = pd.DataFrame(lines, columns=columns)
    return df
```

Transforming data to DataFrame

```python
samplesPath = pd.read_csv("CodeSamples - Consolidado_1.csv")
```

```python
samplesDF = createDataFrame(filterSamples(samplesPath, questions))
samplesDF.to_csv("sampleQuestions.csv", index=False)
samplesDF["path"].drop_duplicates().to_csv("sampleWithQuestions.csv", index=False)
```
